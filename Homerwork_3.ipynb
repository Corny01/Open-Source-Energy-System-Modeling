### Introduction
##### Shared Socioeconomic Pathways
The Shared Socioeconomic Pathways (SSP) are modelled scenarios that show socioeconomic changes up to 2100 depending on different climate policies around the world. There are five different pathways of witch each applies to one possible global climate policy:
+ SSP1: Sustainability (Taking the Green Road)
+ SSP2: Middle of the road
+ SSP3: Regional rivalry (A Rocky Road)
+ SSP4: Inequality (A Road Divided)
+ SSP5: Fossil-Fueled Development (Taking the Highway)
  
Based on these five scenariosthe IPCC Sixth Assessment Report estimates the total global warming in degrees. The report does not estimate the likelihoods of the scenarios.

##### Human Development Index
The Human Development Index (HDI) is a composite statistic used to measure a country's overall achievement in three basic dimensions of human development: health, education, and standard of living. It was developed by the United Nations Development Programme (UNDP) and is published annually in the Human Development Report.  
The three dimensions are measured as followed:
+ Health: life expectancy
+ Education: two indicators 
    + mean years of schooling
    + expected years of schooling
+ standard of living: gross national income per capita
  
The HDI is calculated by combining the normalized scores for each of the three dimensions and its value ranges from 0 to 1. Countries are classified into four tiers based on their HDI score:
+ Very high human development (0.800 and above)
+ High human development (0.700–0.799)
+ Medium human development (0.550–0.699)
+ Low human development (0.549 an below)

### Human Development Index forecast for different socioeconomic Pathway
Crespo Cuaresmaand and Lutz (2016) use the SSPs to forecast the HDI for different countries and continents up to the year 2075. The basis year is 2015. The authors published their results in in the form of data tables. The structure of the data provided by them is shown below.

```python
# transform and import data
# !pip install pyam-iamc
import pyam
import pandas as pd

xls = pd.read_excel("SSP-Extensions_Human_Development_Index_v1.0.xlsx", sheet_name='data')
xls.to_csv('SSP-Extensions_Human_Development_Index_v1.0.csv', index=False, sep=',')

df = pyam.IamDataFrame(data="SSP-Extensions_Human_Development_Index_v1.0.csv")

# show data structure
print(df.head())
```

The regions the data is provided for are shown below in alphabetical order.

```python
df.region
```

##### Relationship between HDI and SSP for Europe
The following graph shows the development of the HDI in Europe for the five different SSPs. The indicator weighted by population is used as it shows best the living conditions for the majority of the people living on the continent.

```python
df.filter(variable="Human Development Index [weighted by Population]", region='Europe (R10)').plot()
```

Unsurprisingly, the HDI is developing best for the SSP1 as the consquences of climate change would be the most harmless ones. More surprising is the positive development of the HDI for the SSP5. Despite the unrestricted use of fossil fuels, the HDI rises sharply until around 2040 before falling only slightly. The worst scenario for the European continent would be SSP3 in the long run. 

##### Comparison to the HDI development in Africa
The following graph shows the development of the HDI in Africa in comparison. Again the indicator weighted by population is used.

```python
df.filter(variable="Human Development Index [weighted by Population]", region='Africa (R10)').plot()
```

One can easily tell that the average HDI in Africa is generally lower than in Europe for all SSPs, especially at the starting point in 2010. On the other hand, the increase in the HDI is stronger for the SSP5 and SSP1 scenarios in particular. The African continent also benefits from the continued use of fossil fuels - even more than in the sustainable scenario. This already shows an initial conflict of interest between the Global North, which would benefit more from strict climate protection plans. It can also be seen that the African continent would suffer greatly from regional rivalry and rising inequality, as the HDI would barely increase in these scenarios until 2075.

##### Comparison to the HDI development in Africa and Europe for SSP1 and SSP4
The following Graph combines the previous results and shows the different HDI developments for Europe an Africa for the SSP1 and SSP4.

```python
regions = ['European Union (R9)', 'Africa (R10)']
scenarios = ['SSP1', 'SSP4']
df.filter(variable="Human Development Index [weighted by Population]", region=regions, scenario=scenarios).plot()
```

It can be seen that the gap between Africa and Europe in the SSP4 would remain largely the same as before. Both curves rise only slightly. For SSP1, on the other hand, the curves of the two continents converge. Social differences between the continents would be reduced. The SSP1 scenario therefore shows that a common climate policy can be positive for the reduction of global social differences.

##### Conclusion
Analysing the HDI forecast for different SSPs and continents shows that there is a strong correlation between the development of a continent and a global climate policy. However, no fundamental advantage of one pathway can be recognised. It can be assumed, however, that a global strategy has a significantly more positive influence on the HDI - be it through the intensification of global markets and the belief in rapid technological progress or through a common restrictive climate policy. Regional rivalries and inequalities are not favourable to either continent.

##### Limitations
One limitation lies in the data itself. The preparation of the data is essential for its informative value. The following chart, which shows the HDI for Asia for the SSP4, illustrates the extent to which the significance of the data can vary depending on how it is processed. The index has a different meaning depending on whether the unweighted average, GDP weighting or population weighting is used.

```python
variables = ["Human Development Index [weighted by Population]", "Human Development Index [unweighted mean]", "Human Development Index [weighted by GDP]"]
df.filter(variable=variables, region='Asia (R5)', scenario='SSP4').plot()
```

The HDI concept covers various areas of a country's development. Nevertheless, its informative value is limited. Concentrating on just three different indicators leaves out many variables. In the above results, it must also be noted that no data is available for some countries in Africa, which could distort the significance of the average values. In general, the focus on the concept of development and developing countries is itself based on ideas and perceptions of the Global North and therefore runs the risk of not properly addressing the needs of the Global South. This is particularly relevant in light of the fact that, as we have seen, a common global strategy is important for positive global development.
